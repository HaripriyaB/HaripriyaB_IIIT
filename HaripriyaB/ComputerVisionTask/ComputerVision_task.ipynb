# Importing necessary libraries

```python
import numpy as np
import pickle
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.constraints import maxnorm
from keras.optimizers import SGD
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.utils import np_utils
from numpy import zeros, newaxis
import csv
```

# Unpickling the given dataset

```python
f=open('train_image.pkl', 'rb')
data = pickle.load(f)
print(data)
f.close
f=open('train_label.pkl', 'rb')
trainlabel = pickle.load(f)
print(trainlabel)
f.close
f=open('test_image.pkl', 'rb')
test = pickle.load(f)
print(test)
f.close
```

# Checking the training labels  

```python
trainlabel=np.asarray(trainlabel)
print(trainlabel)
np.unique(trainlabel)
```

# Converting list to numpy array for easy data analysis and shaping

```python
data=np.asarray(data)
test=np.asarray(test)
```

# Normalizing the dataset for uniform processing and to reduce higher values

```python
data = data.astype('float32')
test = test.astype('float32')
data = data / 255.0
test = test / 255.0
```

# Converting labels to binary vector 

```python
label = np_utils.to_categorical(trainlabel)
num_classes = label.shape[1]
label=np.asarray(label)
```

# Reshaping the datasets to fit in the model

```python
train=[]
for i in range(8000):
    train.append(data[i].reshape(28,28,1))
train=np.asarray(train)
testdata=[]
for i in range(2000):
    testdata.append(test[i].reshape(28,28,1))
testdata=np.asarray(testdata)
```

# Shapes of the datasets 

```python
print(train.shape,testdata.shape,label.shape)
```

# The below model is build by adding few changes and layers of the above simple model.

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(28,28,1), activation='relu'))
model.add(Dropout(0.2))
model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(Dropout(0.2))
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(Dropout(0.2))
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dropout(0.2))
model.add(Dense(1024, activation='relu', kernel_constraint=maxnorm(3)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu', kernel_constraint=maxnorm(3)))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))
#compile
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

# Training of the model is been done 

```python
model.fit(train, label, epochs=25, batch_size=32,verbose=1)
```

# Accuracy of the model is tested with the training dataset 

```python
scores = model.evaluate(train, label, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

# Prediction of test data labels done below

```python
y=model.predict_classes(testdata)
for i in range(len(testdata)):
    print("X=%d, Predicted=%s" % (i, y[i]))
```

# Writing the predicted classes in to a csv file

```python
with open('Haripriya_B.csv', 'w') as f:
    fieldnames = ['image_index', 'class']
    writer = csv.writer(f)
    writer.writerow(fieldnames)
    for i in range(len(testdata)):
        writer.writerow([str(i),str(y[i])])
```
